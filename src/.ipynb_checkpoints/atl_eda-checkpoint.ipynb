**Atlanta Hawks Teammate Networks (1949-2018)**

In this notebook, we are pulling in ATL's nodelist and edgelist to run basic descriptives over time.

```python
# import packages
import os 
import pandas as pd 
import networkx as nx
import seaborn as sns
import matplotlib.pyplot as plt
from networkx.algorithms import community
```

```python
# load node and edgelist 
os.chdir("../data")
nodelist = pd.read_csv("ATL_Network - Nodelist.csv", index_col=1)
edgelist = pd.read_csv("ATL_Network - Edgelist.csv")
edgelist.columns = [col.lower() for col in edgelist]
print(len(nodelist))
print(len(edgelist))
```

```python
nodelist.head()
```

```python
edgelist.head()
```

```python
# count by year from 1946-1979
pre_count_per_year = pd.DataFrame(edgelist.loc[:, '1946':'1975'].sum(skipna = True))
pre_count_per_year = pre_count_per_year.reset_index()
pre_count_per_year = pre_count_per_year.rename(columns = {"index": "year", 0: "edge_count"})
pre_count_per_year['year'] = pre_count_per_year['year'].astype(int)
pre_count_per_year.head()
```

```python
# visualize the counts per year 
sns.set(rc={'figure.figsize':(16,4)})
sns.set_style("white")
pre_yxy_count = sns.barplot(x="year", y="edge_count", data=pre_count_per_year)
plt.title('Pre-Merger Teammate Connections - ATL')
pre_yxy_count.set_xticklabels(pre_yxy_count.get_xticklabels(), rotation=45)
pre_yxy_count
```

```python
# count by year from 1990-2018
post_count_per_year = pd.DataFrame(edgelist.loc[:, '1976':'2018'].sum(skipna = True))
post_count_per_year = post_count_per_year.reset_index()
post_count_per_year = post_count_per_year.rename(columns = {"index": "year", 0: "edge_count"})
post_count_per_year['year'] = post_count_per_year['year'].astype(int)
post_count_per_year.head()
```

```python
# visualize the counts per year 
sns.set(rc={'figure.figsize':(16,4)})
sns.set_style("white")
post_yxy_count = sns.barplot(x="year", y="edge_count", data=post_count_per_year)
plt.title('Post-Merger Teammate Connections - ATL')
post_yxy_count.set_xticklabels(post_yxy_count.get_xticklabels(), rotation=45)
post_yxy_count
```

```python
# create a weighted edgelist 
wtd_edgelist = edgelist[["source","target", "onset", "terminus", "weight"]].sort_values(by=['weight'], ascending=False)
wtd_edgelist
```
